###Checking Hash Logic

```python
import hashlib
```

```python
source_id = []
hashList = []
source_id_set = set()
hashIdSet = set()
for i in sashi_cve['Vul_id']:
  seed_text = "sashi_cve_"+i
  source_id.append(seed_text)
  source_id_set.add(seed_text)
  seed_text = seed_text.encode('utf-8')
  hashId = hashlib.sha256(seed_text).hexdigest()
  hashList.append(hashId)
  hashIdSet.add(hashId)
print("-----------------------------------------------------------------------------------------------------------------------")
print(len(source_id))
print(len(source_id_set))
print(len(source_id)-len(source_id_set))
print("############Hash Details################")
print(len(hashList))
print(len(hashIdSet))
print(len(hashList)-len(hashIdSet))
print("-----------------------------------------------------------------------------------------------------------------------")
for i in sashi_cve_c['CVE ID']:
  seed_text = "sashi_cve_c_"+i
  source_id.append(seed_text)
  source_id_set.add(seed_text)
  seed_text = seed_text.encode('utf-8')
  hashId = hashlib.sha256(seed_text).hexdigest()
  hashList.append(hashId)
  hashIdSet.add(hashId)
print("-----------------------------------------------------------------------------------------------------------------------")
print(len(source_id))
print(len(source_id_set))
print(len(source_id)-len(source_id_set))
print("############Hash Details################")
print(len(hashList))
print(len(hashIdSet))
print(len(hashList)-len(hashIdSet))
print("-----------------------------------------------------------------------------------------------------------------------")
for i in karthic_cve_v['_id']:
  seed_text = "karthic_cve_v"+i
  source_id.append(seed_text)
  source_id_set.add(seed_text)
  seed_text = seed_text.encode('utf-8')
  hashId = hashlib.sha256(seed_text).hexdigest()
  hashList.append(hashId)
  hashIdSet.add(hashId)
print("-----------------------------------------------------------------------------------------------------------------------")
print(len(source_id))
print(len(source_id_set))
print(len(source_id)-len(source_id_set))
print("############Hash Details################")
print(len(hashList))
print(len(hashIdSet))
print(len(hashList)-len(hashIdSet))
print("-----------------------------------------------------------------------------------------------------------------------")
```

```python
headers = dict()
headers[
    "User-Agent"
] = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36"
df = pd.DataFrame(columns=['Vul_id','Summary', 'CVSS_sev'])
```

```python
pages = np.arange(1, 300, 20)
for page in pages:
    url = "https://nvd.nist.gov/vuln/search/results?isCpeNameSearch=false&query=hp&results_type=overview&form_type=Basic&search_type=all&startIndex" + str(page)
    results = requests.get(url, headers=headers)
    soup = BeautifulSoup(results.text, "html.parser")
    table = soup.find('table', class_='table table-striped table-hover')
    print(page)
    # Collecting Ddata
    for row in table.tbody.find_all('tr'):    
    # Find all data for each column
     h = row.find_all('th')
     columns = row.find_all('td')

     if(h != []):
        Vul_id = h[0].text.strip()
     if(columns != []):
        
        Summary = columns[0].text.strip()
        CVSS_sev = columns[1].text.strip()

        df = df.append({'Vul_id': Vul_id,  'Summary': Summary , 'CVSS_sev' : CVSS_sev }, ignore_index=True)
```
